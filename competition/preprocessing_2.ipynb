```python
import numpy as np
import pandas as pd
```

```python
DATA_PATH = './data/'

raw_train = pd.read_csv(f'{DATA_PATH}pre_train_1.csv')
raw_test = pd.read_csv(f'{DATA_PATH}pre_test_1.csv')

raw_train.set_index('id', inplace=True)
raw_test.set_index('id', inplace=True)

raw_train.shape, raw_test.shape
```

```python
def convert_dtype(pDf):
  df = pDf.copy()

  int_df = df.select_dtypes(include=['int'])
  int_df_str = int_df.applymap(str)

  bool_df = df.select_dtypes(include=['bool'])
  bool_df_int = bool_df.applymap(int)

  object_df = df.select_dtypes(include=['object'])

  pre_df =  pd.concat([int_df_str, bool_df_int, object_df], axis=1)
  int_cols = [ 
    'person_attribute_a_1', 'person_attribute_b', 'person_prefer_e', 'contents_attribute_e'
  ]
  pre_df[int_cols] = pre_df[int_cols].astype(int)

  return pre_df
```

```python
pre_train_2 = convert_dtype(raw_train)
pre_test_2 = convert_dtype(raw_test)

pre_train_2.shape, pre_test_2.shape
```

```python
pre_train_2.info()
```

```python
len(pre_train_2.select_dtypes(include=['object']).columns)
```

```python
obj_cols = list(pre_train_2.select_dtypes(include=['object']).columns)
obj_cols.remove('contents_open_dt')
obj_cols.remove('person_rn')
obj_cols.remove('contents_rn')
obj_cols.remove('target')
len(obj_cols)
```

```python
pre_train_2[obj_cols] = 's'+pre_train_2[obj_cols]
pre_test_2[obj_cols] = 's'+pre_test_2[obj_cols]
```

```python
pre_train_2.to_csv(f'{DATA_PATH}pre_train_2.csv', index=True)
pre_test_2.to_csv(f'{DATA_PATH}pre_test_2.csv', index=True)
```
